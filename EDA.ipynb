```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape') # to avoid error
```

```python
# drop unrelated/blank columns
df.drop(['Status','unnamed1'],axis = 1, inplace = True)
```

```python
pd.isnull(df)
```

```python
#check for null vales - sum
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True) # to save the deleted thing we use inplace = True
#df.dropna(inplace = True) and df = df.dropna() both are same
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
# Exploratory Data Analysis
# Gender
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_gen = df.groupby(['Gender'], as_index = False) [ 'Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender',y = 'Amount', data = sales_gen)
```

```python
# Age
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index = False) [ 'Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group',y = 'Amount', data = sales_age)
```

```python
# State
# total orders from top 10 states

sales_state = df.groupby(['State'],as_index = False) ['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = sales_state , x = 'State', y = 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = sales_state , x = 'State', y = 'Amount')
```

```python
# Marital Status

ax = sns.countplot(data = df, x= 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

sales_marriage = df.groupby(['Marital_Status','Gender'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(data = sales_marriage , x = 'Marital_Status', y = 'Amount',hue = 'Gender')
```

```python
# Occupation

sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_occupation = df.groupby(['Occupation'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(17,5)})
sns.barplot(data = sales_occupation , x = 'Occupation', y = 'Amount')
```

```python
# Product Category
sns.set(rc = {'figure.figsize':(25,10)})
ax = sns.countplot(data = df, x= 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_product_category = df.groupby(['Product_Category'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,10)})
sns.barplot(data = sales_product_category , x = 'Product_Category', y = 'Amount')
```

```python
# Top 10 most sold products
sales_product_ID = df.groupby(['Product_ID'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,10)})
sns.barplot(data = sales_product_ID , x = 'Product_ID', y = 'Amount')
```
